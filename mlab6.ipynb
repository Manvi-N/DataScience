```python
df.isna().sum()
```

```python
oht = pd.get_dummies(df)
```

```python
frequent_item_sets = apriori(oht, min_support=0.4, use_colnames=True)
```

```python
association_rules_df = association_rules(frequent_item_sets, metric='confidence', min_threshold=0.8)
```
