```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X=data[['Annual Income (k$)','Spending Score (1-100)']]
se=[]
for k in range(1,21):
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X)
    se.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,21),se,marker='o')
plt.xlabel('No. of Clusters')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow')
plt.show()
```

```python
optimal_clusters=4
kmeans=KMeans(n_clusters=optimal_clusters,random_state=42)
kmeans.fit(X)
```

```python
data['Cluster']=kmeans.labels_
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',hue='Cluster',data=data,palette='colorblind')
plt.title('K-means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score9(1-100)')
plt.show()
```

```python
optimal_clusters=3
kmeans=KMeans(n_clusters=optimal_clusters,random_state=42)
kmeans.fit(X)
```

```python
data['Cluster']=kmeans.labels_
plt.figure(figsize=(10,8))
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',hue='Cluster',data=data,palette='colorblind')
plt.title('K-means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score9(1-100)')
plt.show()
```

```python
q1=data['Annual Income (k$)'].quantile(0.45)
q3=data['Annual Income (k$)'].quantile(0.65)
iqr=q3-q1
lw=q1-(1.5*iqr)
uw=q3+(1.5*iqr)
df=data[(data['Annual Income (k$)']>=lw) & (data['Annual Income (k$)']<=uw)]
#plt.figure(figsize=(10,8))
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',hue='Cluster',data=df,palette='colorblind')
```
